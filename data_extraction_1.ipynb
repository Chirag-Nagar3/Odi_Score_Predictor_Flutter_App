```python
from yaml import safe_load
```

```python
from tqdm import tqdm
```

```python
import pandas as pd
import os
import numpy as np
```

```python
filename = []
for file in os.listdir('odis_male'):
    filename.append(os.path.join('odis_male',file))
```

```python
filename[0:5]
```

```python
final_df = pd.DataFrame()
counter = 1
for file in tqdm(filename):
    with open (file, 'r') as f:
        df = pd.json_normalize(safe_load(f))
        df['match_id'] = counter
        final_df = final_df.append(df)
        counter+=1
        
final_df
```
